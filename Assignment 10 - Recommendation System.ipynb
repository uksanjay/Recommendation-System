```python
book_df_new = book_df.pivot_table(values= 'Book_Rating',index='User_ID' , columns= 'Book_Title')
# Creating pivot_table for values.
```

```python
book_df_new.fillna(0,inplace=True)
# Replacing NAN with 0 for similarity calculation
```

```python
book_df_new.head()
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(book_df_new , metric = 'cosine')
# calculating user similarity based on cosine metric.
```

```python
user_sim
```

```python
np.fill_diagonal(user_sim , 0)
# Replacing diagonal values with 0.
```

```python
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df
# DF of similarity between users , 0 being least similar and 1 being highly similar.
```

```python
user_sim_df.index = book_df.User_ID.unique()
user_sim_df.columns = book_df.User_ID.unique()
```

```python
user_sim_df.head()
```

```python
X = user_sim_df[user_sim_df.values > 0]
X
# Users with similarities.
```

```python
Y = user_sim_df[user_sim_df.values > .5 ]
Y.reset_index()
# Users with similarity > 0.5
```

```python
user = []
for i in range(0,42):
    x = Y.iloc[i ,: ].idxmax()
    user.append(x)
        
```

```python
Similar_users = pd.DataFrame()
Similar_users['User_1'] = pd.Series(user)
Similar_users['User_2'] = Y.index
```

```python
Similar_users
# Users with similarity > 0.5
```
